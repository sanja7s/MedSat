# Define Data Inputs to the Models

```python
controls = [c for c in data.columns if c.startswith("c_")]
environment = [c for c in data.columns if c.startswith("e_")]
outcomes = [c for c in data.columns if c.startswith("o_")]

features = controls + environment
```

```python
outcome = 'o_diabetes_quantity_per_capita'
```

# VIF Colinearity

```python
X = data[features].dropna()

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

# Functions

```python
def LR(data=data, index_column="geography code", label=outcome, features_columns=features):
	
	data2 = data.copy()
	
	target = data2[[index_column, label]]
	features = data2[features_columns]
	
	X = features.values
	y = target[label].values
	
	X = sm.add_constant(X)  
	model = sm.OLS(y,X, missing='drop')	
	results = model.fit()
	
	print ("LABEL ", label)
	print_model = results.summary(xname=['intercept']+features_columns, yname =  outcome)
	print (print_model)
	
	return results.rsquared_adj
```

# Apply Models

## Opioids

```python
outcome = "o_opioids_quantity_per_capita"
LR(data=data, index_column="geography code", label=outcome, features_columns=features)
```

```python
outcome = "o_OME_per_capita"
LR(data=data, index_column="geography code", label=outcome, features_columns=features)
```

## Diabetes

```python

outcome = "o_diabetes_quantity_per_capita"
LR(data=data, index_column="geography code", label=outcome, features_columns=features)
```

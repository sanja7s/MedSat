```python
def concat_files(input_folder, output_filename, column_name=None):
  DF = pd.DataFrame()
  for f in os.listdir(data_folder + input_folder):
    if f.endswith('.csv'):
      if 'whole_England' in f:
        continue
      df = pd.read_csv(data_folder + input_folder + f)
      print (input_folder, f, len(df), len(df.drop_duplicates()), len(set(df['LSOA21CD'])))
      if len(DF):
        print (len(set(df['LSOA21CD']).intersection(set(DF['LSOA21CD']))))
      DF = pd.concat([DF,df])
      print(len(DF), len(DF.drop_duplicates()), len(set(DF['LSOA21CD'])))
  print(len(DF), len(DF.drop_duplicates()))
  if 'mean' in DF.columns and column_name!=None:
    DF = DF.rename(columns={'mean':column_name})
  DF.to_csv(data_folder + input_folder + output_filename, index=None)
  return DF
```

# Land Cover asks for a special preprocessing

```python
input_folder = 'ESA-WorldCover/lsoa/2020/'
output_filename = 'whole_England_land_cover.csv'

tmp_orig =  concat_files(input_folder, output_filename)
```

```python
tmp = tmp_orig.copy()

# Define the keys you want to extract
keys = ['10', '20', '30', '40', '50', '60', '70', '80', '90', '95', '100']
keys2 = ['_'+k for k in keys]

rename_columns = {
"10":		"Tree cover",
"20":		"Shrubland",
"30":		"Grassland",
"40":		"Cropland",
"50":		"Built-up",
"60":		"Bare / sparse vegetation",
"70":		"Snow and ice",
"80":		"Permanent water bodies",
"90":		"Herbaceous wetland",
"95":		"Mangroves",
"100":	"Moss and lichen"
}

# Convert single quotes to double quotes
tmp['normalized_histogram'] = tmp['normalized_histogram'].str.replace("'", "\"")
tmp['histogram'] = tmp['histogram'].str.replace("'", "\"")

# Convert the strings to dictionaries
tmp['normalized_histogram'] = tmp['normalized_histogram'].apply(json.loads)
tmp['histogram'] = tmp['histogram'].apply(json.loads)

# Define a function to extract the values for each key
def extract_values(row):
    return [row.get(key, None) for key in keys]

# Apply the function to create separate columns for each key
tmp[keys] = tmp['normalized_histogram'].apply(extract_values).apply(pd.Series)

# we do not need the histogram column anymore
del tmp['histogram']
del tmp['normalized_histogram']

# let us use the names for the land classes instead of codes
tmp.rename(columns=rename_columns, inplace=True)

# TODO check with LSOA size
# tmp[keys2] = tmp['histogram'].apply(extract_values).apply(pd.Series)

tmp.head()
```

```python
master_df = master_df.merge(tmp, on=['LSOA21NM',	'LSOA21CD'])
len(master_df)
```

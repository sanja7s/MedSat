```python
import networkx as nx
import pandas as pd
from code.matching.utils import *
import json
```

```python
graph = nx.read_gexf("code/mappings/drug_association_graph.gexf")
chem = pd.read_csv('code/mappings/CHEM_MASTER_MAP.csv')
```

```python
def findDrugsForDisease_2(Graph, Disease, BNF_Chem ):#,threshProb):
    chem_dict = makeChemDict(BNF_Chem)
    drugs = {}
    for e in Graph.edges(data=True):
        if (cleanStringofUTF(e[1]).lower().find(Disease.lower()) >=0) or (cleanStringofUTF(e[0]).lower().find(Disease.lower()) >= 0) :
            drugNode = ''
            matchedDisease = ''
            if Graph.nodes[e[0]]['type'] == 'symptom':
                drugNode = e[1]
                matchedDisease = e[0]
            else:
                drugNode = e[0]
                matchedDisease = e[1]
            drugs[Graph.nodes[drugNode]['Id']] = {}
            drugs[Graph.nodes[drugNode]['Id']]['name'] = drugNode
            drugs[Graph.nodes[drugNode]['Id']]['matched_disease'] = matchedDisease
            drugs[Graph.nodes[drugNode]['Id']]['disease'] = Disease
    enrichdrugs(chem_dict,drugs)
    return drugs

def dumpDrugs(self, matchedDrugs):
    drug_map_dict = {'BNF_code':[], 'Drug_name':[] , 'Mapped_Condition': []}
    for k in matchedDrugs:
        drug_map_dict['BNF_code'].append(k)
        drug_map_dict['Drug_name'].append(matchedDrugs[k]['chemName'])
        drug_map_dict['Mapped_Condition'].append(matchedDrugs[k]['disease'])
    drug_map_df = pd.DataFrame.from_dict(drug_map_dict) 
    return drug_map_df
```

```python
findDrugsForDisease_2(graph , "asthma" , chem)
```

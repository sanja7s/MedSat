```python
def concat_files(input_folder, output_filename, column_name=None):
  DF = pd.DataFrame()
  for f in os.listdir(data_folder + input_folder):
    if f.endswith('.csv'):
      if 'whole_England' in f:
        continue
      df = pd.read_csv(data_folder + input_folder + f)
      print (input_folder, f, len(df), len(df.drop_duplicates()), len(set(df['LSOA21CD'])))
      if len(DF):
        print (len(set(df['LSOA21CD']).intersection(set(DF['LSOA21CD']))))
      DF = pd.concat([DF,df])
      print(len(DF), len(DF.drop_duplicates()), len(set(DF['LSOA21CD'])))
  print(len(DF), len(DF.drop_duplicates()))
  if 'mean' in DF.columns and column_name!=None:
    DF = DF.rename(columns={'mean':column_name})
  DF.to_csv(data_folder + input_folder + output_filename, index=None)
  return DF
```

```python
the_year = 2022
```

# Sentinel-5 NO2

```python
input_folder = 'S5/lsoa/{}/'.format(the_year)
output_filename = 'whole_England_NO2.csv'

master_df = concat_files(input_folder, output_filename, column_name='NO2')
len(master_df)
# master_df[['LSOA21NM',	'LSOA21CD']].drop_duplicates()
```

```python
input_folder = 'OMI/lsoa/{}/'.format(the_year)
output_filename = 'whole_England_ozone.csv'

tmp = concat_files(input_folder, output_filename, column_name='ozone')
master_df = master_df.merge(tmp, on=['LSOA21NM',	'LSOA21CD'])
# len(master_df)
```
